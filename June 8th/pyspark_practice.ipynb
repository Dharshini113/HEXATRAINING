```python
from pyspark.sql import SparkSession

spark = SparkSession.builder \
    .appName("PySpark Day 1") \
    .getOrCreate()
```

```python
data = [
    (101, "Rahul Sharma", "Data Engineering", 75000, "Hyderabad"),
    (102, "Priya Reddy", "AI Engineering", 85000, "Bangalore"),
    (103, "Amit Kumar", "Data Engineering", 65000, "Mumbai"),
    (104, "Sneha Patel", "Data Science", 95000, "Chennai"),
    (105, "Farhan Ali", "Cloud Engineering", 80000, "Delhi")
]
```

```python
columns = [
    "employee_id",
    "name",
    "department",
    "salary",
    "city"
]
```

```python
df = spark.createDataFrame(data, columns)
df.show()
```

```python
df = df.withColumn(
    "salary_category",
    when(col("salary") >= 85000, "High")
    .when(col("salary") >= 75000, "Medium")
    .otherwise("Low")
)
df.show()
```

```python
report_df = (
    df.withColumn("bonus", df.salary * 0.10)
      .withColumn("annual_salary", df.salary * 12)
      .select(
          "employee_id",
          "name",
          "department",
          "salary",
          "bonus",
          "annual_salary"
      )
)

report_df.show()
```
