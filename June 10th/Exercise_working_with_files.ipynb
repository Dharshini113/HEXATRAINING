```python
from pyspark.sql import SparkSession
spark = SparkSession.builder.appName("working with files").getOrCreate()
```

working with CSV file

```python
from google.colab import files
files.upload ()
```

```python
df = spark.read.csv("flightbooking.csv",header = True , inferSchema = True)
df.show()
```

```python
df.printSchema()
```

```python
# 1
df.show()
```

```python
# 2
df.select("passenger_name", "from_city","to_city").show()
```

```python
# 3
df.filter(df.status == "Confirmed").show()
```

```python
# 4
df.filter(df.status == "Cancelled").show()
```

```python
# 5
df.filter(df.status == "Pending").show()
```

```python
# 6
df.filter(df.ticket_price > 20000).show()
```

```python
# 7
df.filter(df.to_city .isin("Singapore","Dubai","London")).show()
```

```python
# 8
df.count()
```

```python
# 9
df.groupBy("airline").count().show()
```

```python
# 10
df.groupBy("status").count().show()
```

```python
# 11
from pyspark.sql.functions import *
df.filter(df.status == "Confirmed").select(sum(df.ticket_price)).show()
```

```python
# 12
df.groupBy("airline").avg("ticket_price").show()
```

```python
# 13
df.select(max("ticket_price")).show()
```

```python
# 14
df.select(min("ticket_price")).show()
```

```python
# 15
df.orderBy(df.ticket_price.desc()).show()
```

```python
# 16
df1 = df.withColumn("tax",df.ticket_price * 0.05)
df1.show()
```

```python
df1.show()
```

```python
# 17
df2 = df1.withColumn("final_price" , df1.ticket_price + df1.tax)
df2.show()
```

```python
# 18
df3 = df.withColumn("price_category", when(df.ticket_price >= 30000 ,"Premium")
.when(df.ticket_price >=10000 ,"Standard").otherwise("Budget"))
df3.show()
```

```python
# 19
df3.groupBy("price_category").count().show()
```

```python
df.filter(df.status == "Confirmed").write.mode("overwrite").parquet("confirmed_flights.parquet")
```

```python
spark.read.parquet("confirmed_flights.parquet").show()
```

working with JSON file

```python
from google.colab import files
files.upload ()
```

```python
hdf = spark.read.option( "multiline", "true" ).json("hotels.json")
hdf.show()
```

```python
hdf.printSchema()
```

```python
# 1
hdf.show()
```

```python
# 2
hdf.select("hotel_name","city","rating").show()
```

```python
# 3
hdf.filter(hdf.category == "Luxury").show()
```

```python
# 4
hdf.filter(hdf.rating > 4.5).show()
```

```python
# 5
hdf.filter(hdf.rooms_available > 15).show()
```

```python
# 6
hdf.filter(hdf.price_per_night > 10000).show()
```

```python
# 7
hdf.filter(hdf.city.isin("Dubai","London")).show()
```

```python
# 8
hdf.filter(hdf.contact.phone.isNull()).show()
```

```python
# 9
hdf.filter(hdf.contact.email.isNull()).show()
```

```python
# 10
hdf.select("hotel_name","contact.phone","contact.email").show()
```

```python
# 11
hdf.filter(array_contains(hdf.amenities,"wifi")).show()
```

```python
# 12
hdf.filter(array_contains(hdf.amenities,"spa")).show()
```

```python
# 13
hdf.groupBy("city").count().show()
```

```python
# 14
hdf.groupBy("category").count().show()
```

```python
# 15
hdf.groupBy("category").avg("rating").show()
```

```python
# 16
hdf.groupBy("city").avg("price_per_night").show()
```

```python
# 17
hdf.select(max("price_per_night")).show()
```

```python
# 18
hdf = hdf.withColumn(
    "total_potential_revenue",
    hdf.rooms_available * hdf.price_per_night
)

hdf.show()
```

```python
# 19
hdf.orderBy(hdf.rating.desc()).show()
```

```python
hdf.select(
    "hotel_id",
    "hotel_name",
    "city",
    "category",
    "rating",
    "rooms_available",
    "price_per_night",
    "contact.phone",
    "contact.email",
    "total_potential_revenue"
).show()
```

```python
# 20
hdf.write.mode("overwrite").parquet("hotels_flattened.parquet")
```

```python
spark.read.parquet("hotels_flattened.parquet").show()
```
