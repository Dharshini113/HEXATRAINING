Window Function Exercises

```python
from pyspark.sql.window import *
doctor_revenue = doctor_visit.groupBy(
    "doctor_id",
    "doctor_name",
    "specialization",
    "city"
).agg(
    sum("bill_amount").alias("revenue")
)
rev_win = Window.orderBy(desc("revenue"))
```

```python
doctor_revenue.withColumn(
    "rank",
    rank().over(rev_win)
).show()

doctor_revenue.withColumn(
    "dense_rank",
    dense_rank().over(rev_win)
).show()

doctor_revenue.withColumn(
    "row_number",
    row_number().over(rev_win)
).show()

doctor_revenue.withColumn(
    "rank",
    rank().over(rev_win)
).filter(
    col("rank")==1
).show()

doctor_revenue.withColumn(
    "rank",
    rank().over(rev_win)
).filter(
    col("rank")<=1
).show()
```

```python
spec_window = Window.partitionBy("specialization").orderBy(col("revenue").desc())

doctor_revenue.withColumn(
    "rank",
    rank().over(spec_window)
).filter(col("rank")==1).show()

doctor_revenue.withColumn(
    "rank",
    rank().over(spec_window)
).filter(col("rank")<=2).show()
```

```python
doctor_revenue.withColumn(
    "running_revenue",
    sum("revenue").over(rev_win)
).show()

doctor_revenue.withColumn(
    "prev_revenue",
    lag("revenue").over(rev_win)
).show()

doctor_revenue.withColumn(
    "next_revenue",
    lead("revenue").over(rev_win)
).show()

doctor_revenue.withColumn(
    "difference",
    col("revenue")-lag("revenue").over(rev_win)
).show()

doctor_revenue.withColumn(
    "difference",
    lead("revenue").over(rev_win) - col("revenue")
).show()
   
```

```python

city_window = Window.partitionBy("city").orderBy(col("revenue").desc())

doctor_revenue.withColumn(
    "rank",
    rank().over(city_window)
).filter(
    col("rank")==1
).show()
```
