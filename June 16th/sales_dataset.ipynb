
Window Functions

```python
sales_df = df.withColumn("revenue",col("quantity") * col("price"))
window_spec = Window.orderBy(col("revenue").desc())
```

```python
sales_df.withColumn("row_number",row_number().over(window_spec)).show()
```

```python
sales_df.withColumn("rank", rank().over(window_spec)).show()
```

```python
sales_df.withColumn("dense_rank",dense_rank().over(window_spec)).show()
```

```python
sales_df.orderBy(col("revenue").desc()).limit(5).show()
sales_df.orderBy(col("revenue").desc()).limit(1).show()
```

```python
city_w = Window.partitionBy("city").orderBy(col("revenue").desc())

sales_df.withColumn(
    "rn",
    row_number().over(city_w)
).filter(col("rn") == 1).show()
```

```python
category_win = Window.partitionBy("category").orderBy(col("revenue").desc())

sales_df.withColumn(
    "rn",
    row_number().over(category_win)
).filter(col("rn") == 1).show()
```

```python
running_window = Window.orderBy("order_id")

sales_df.withColumn(
    "running_total",
    sum("revenue").over(running_window)
).show()
```

```python
sales_df.withColumn(
    "previous_revenue",
    lag("revenue",1).over(window_spec)
).show()

sales_df.withColumn(
    "next_revenue",
    lead("revenue",1).over(window_spec)
).show()
```

```python
sales_df.withColumn(
    "previous_revenue",
    lag("revenue").over(window_spec)
).withColumn(
    "comparison",
    col("revenue") - col("previous_revenue")
).show()
```

```python
sales_df.withColumn(
    "revenue_difference",
    col("revenue") - lag("revenue").over(window_spec)
).show()
```

```python
sales_df.withColumn(
    "rank",
    row_number().over(city_w)
).filter( col("rank") <= 2).show()
```

```python
sales_df.withColumn(
    "rank",
    row_number().over(category_win)
).filter(col("rank") <= 3).show()
```

```python
sales_df.withColumn(
    "rank",
    dense_rank().over(window_spec)
).select(
    "order_id",
    "product",
    "city",
    "revenue",
    "rank"
).orderBy("rank").show()
```

Mini ETL Exercises

```python
source = df
```

```python
transformed = source.withColumn(
    "total_amount",
    col("quantity") * col("price")
)
```

```python
completed= transformed.filter(
    col("status")=="Completed"
)
```

```python
revenue = completed.withColumn(
    "revenue",
    col("total_amount")
)
```

```python
city_report = revenue.groupBy("city").agg(sum("revenue").alias("city_revenue"))
```

```python
category_report = revenue.groupBy("category").agg(sum("revenue").alias("category_revenue"))
```

```python
spark.sql("SHOW CATALOGS").show(truncate=False)
```

```python
spark.sql("SHOW SCHEMAS IN hexa_databricks_7405612649268533").show(truncate=False)
```

```python
revenue.write.mode(
    "overwrite"
).saveAsTable(
    'hexa_databricks_7405612649268533.default.sales_report'
)
```

```python
parquet_df = spark.table("default.sales_report")

display(parquet_df)
```

```python
parquet_df.createOrReplaceTempView(
    "sales_parquet"
)
```

```python
%sql
select * from sales_parquet;
```

```python
final_report = spark.sql("""
SELECT
    city,
    category,
    COUNT(*) AS total_orders,
    SUM(quantity) AS total_quantity,
    SUM(revenue) AS total_revenue,
    AVG(price) AS avg_price
FROM sales_parquet
GROUP BY city, category
ORDER BY total_revenue DESC
""")

display(final_report)
```
