Part 1: Ingestion

```python
stores_df = spark.read.csv("stores.csv",header=True,inferSchema=True)

products_df = spark.read.csv("products.csv",header=True,inferSchema=True)

inventory_df = spark.read.csv("inventory.csv",header=True,inferSchema=True)

sales_df = spark.read.csv("sales.csv",header=True,inferSchema=True)

suppliers_df = spark.read.option("multiline","true").json("suppliers.json")
```

```python
stores_df.printSchema()
products_df.printSchema()
inventory_df.printSchema()
sales_df.printSchema()
suppliers_df.printSchema()
```

```python
print("Count of Stores:", stores_df.count())
print("Count of Products:", products_df.count())
print("Count of Inventory:", inventory_df.count())
print("Count of Sales:", sales_df.count())
print("Count of Suppliers:", suppliers_df.count())
```

```python
stores_df.write.mode("overwrite").parquet("bronze/stores")
products_df.write.mode("overwrite").parquet("bronze/products")
inventory_df.write.mode("overwrite").parquet("bronze/inventory")
sales_df.write.mode("overwrite").parquet("bronze/sales")
suppliers_df.write.mode("overwrite").parquet("bronze/suppliers")
```

Part 2: Data Cleaning

```python
from pyspark.sql.functions import *
products_df.filter(col("supplier_id").isNull()).show()
```

```python
inventory_df.filter(col("stock_quantity").isNull()).show()
```

```python
sales_df.filter(col("sale_amount").isNull()).show()
```

```python
sales_df.filter(col("payment_mode").isNull()).show()
```

```python
inventory_clean = inventory_df.fillna({
    "stock_quantity": 0
})
inventory_clean.show()
```

```python
sales_clean = sales_df.fillna({
    "sale_amount": 0
})

sales_clean.show()
```

```python
sales_clean = sales_clean.fillna({
    "payment_mode": "Not Provided"
})
sales_clean.show()
```

```python
products_clean = products_df.fillna({
    "supplier_id": "UNKNOWN"
})
products_clean.show()
```

```python
products_clean= products_clean.withColumn(
    "data_quality_status",
    when(col("supplier_id") == "UNKNOWN", "Missing")
    .otherwise("Valid")
)
products_clean.show()
```

```python
inventory_clean = inventory_clean.withColumn(
    "data_quality_status",
    when(col("stock_quantity") == 0, "Missing")
    .otherwise("Valid")
)
inventory_clean.show()
```

```python
sales_clean = sales_clean.withColumn(
    "data_quality_status",
    when(
        (col("sale_amount") == 0) |
        (col("payment_mode") == "Not Provided"),"Missing"
    ).otherwise("Valid")
)
sales_clean.show()
```

```python
products_clean.write.mode("overwrite").parquet("silver/products")
inventory_clean.write.mode("overwrite").parquet("silver/inventory")
sales_clean.write.mode("overwrite").parquet("silver/sales")
```

Part 5: Transformations

```python
inventory_products = inventory_products.withColumn(
    "stock_status",
    when(
        col("stock_quantity") <= col("reorder_level"),
        "Reorder Required"
    ).otherwise("Sufficient Stock")
)
inventory_products.show()
```

```python
products_clean = products_clean.withColumn(
    "price_category",
    when(col("unit_price") >= 50000, "Premium")
    .when(col("unit_price") >= 10000, "Standard")
    .otherwise("Budget")
)
products_clean.show()
```

```python
sales_clean = sales_clean.withColumn(
    "revenue_category",
    when(col("sale_amount") >= 50000, "High Revenue")
    .when(col("sale_amount") >= 15000, "Medium Revenue")
    .otherwise("Low Revenue")
)

sales_clean.show()
```

```python
sales_clean = sales_clean.withColumn(
    "month",
    month(col("sale_date"))
)
sales_clean.show()
```

```python
sales_clean = sales_clean.withColumn(
    "year", year(col("sale_date"))
)
sales_clean.show()
```

```python
inventory_products = inventory_products.withColumn(
    "inventory_value",col("stock_quantity") * col("unit_price")
)
inventory_products.show()
```

```python
suppliers_flat = suppliers_flat.withColumn(
    "supplier_quality",
    when(col("rating") >= 4.5, "Excellent")
    .when(col("rating") >= 4.0, "Good")
    .otherwise("Average")
)
suppliers_flat.show()
```

Part 7: Window Functions

```python
from pyspark.sql.window import *

product_rev = retail_df.groupBy(
    "product_name"
).agg(sum("sale_amount").alias("revenue"))

w = Window.orderBy(desc("revenue"))

product_rev.withColumn(
    "rank",
    rank().over(w)
).show()
```

```python
store_rev = retail_df.groupBy(
    "store_name"
).agg(sum("sale_amount").alias("revenue"))

store_rev.withColumn(
    "rank",
    rank().over(w)
).show()
```

```python
wc = Window.partitionBy("category").orderBy(col("revenue").desc())

category_rev= retail_df.groupBy(
    "category",
    "product_name"
).agg(
    sum("sale_amount").alias("revenue")
)

category_rev.withColumn(
    "rank",
    rank().over(wc)
).show()
```

```python
category_rev.withColumn(
    "rank",
    rank().over(wc)
).filter(col("rank")==1).show()
```

```python
category_rev.withColumn(
    "rank",
    rank().over(wc)
).filter(col("rank")<=3).show()
```

```python
ws = Window.partitionBy("state").orderBy(desc("revenue"))
state_rev = retail_df.groupBy(
    "state",
    "store_name"
).agg(sum("sale_amount").alias("revenue"))

state_rev.withColumn(
    "rank",
    rank().over(ws)
).filter(col("rank")==1).show()
```

```python
wr = Window.orderBy("sale_date")
sales_clean.withColumn(
    "running_total",
    sum("sale_amount").over(wr)
).show()
```

```python
sales_clean.withColumn(
    "previous_day_sales",
    lag("sale_amount").over(wr)
).show()
```

```python
sales_clean.withColumn(
    "next_sale_amount",
    lead("sale_amount").over(wr)
).show()
```

```python
sales_clean.withColumn(
    "previous_sale",
    lag("sale_amount").over(wr)
).filter(
    col("sale_amount") > col("previous_sale")
).show()
```
