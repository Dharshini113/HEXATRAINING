Part 2: Data Cleaning

```python
from pyspark.sql.functions import *
customers_df.filter(col("plan_id").isNull()).show()
```

```python
usage_clean = usage_df.fillna({"data_used_gb":0})
```

```python
payments_clean = payments_df.fillna({
    "amount_paid":0,
    "payment_mode":"Not Provided"
})
```

```python
customers_clean = customers_df.fillna({
    "plan_id":"UNKNOWN"
})
```

```python
customers_clean = customers_clean.withColumn(
    "data_quality_status",
    when(col("plan_id")=="UNKNOWN","Missing").otherwise("Valid")
)

usage_clean = usage_clean.withColumn(
    "data_quality_status",
    when(col("data_used_gb")==0,"Missing").otherwise("Valid")
)

payments_clean = payments_clean.withColumn(
    "data_quality_status",
    when(col("amount_paid")==0,"Missing").otherwise("Valid")
)
```

Part 5: Transformations

```python
full_df = full_df.withColumn(
    "usage_category",
    when(col("data_used_gb") >= 70, "Heavy User")
    .when(col("data_used_gb") >= 30, "Medium User")
    .otherwise("Low User")
)
```

```python
full_df = full_df.withColumn(
    "payment_category",
    when(col("amount_paid") >= 1000, "High Payment")
    .when(col("amount_paid") >= 500, "Medium Payment")
    .otherwise("Low Payment")
)
```

```python
full_df = full_df.withColumn(
    "churn_risk",
    when(
        (col("status")=="Inactive") |
        (col("payment_status")!="Success"),"High Risk"
    )
    .when(col("data_used_gb") < 15,"Medium Risk")
    .otherwise("Low Risk")
)
```

```python
full_df = full_df.withColumn(
    "over_usage_gb",
    col("data_used_gb") - col("data_limit_gb")
)
```

```python
full_df = full_df.withColumn(
    "over_usage_flag",
    when(col("over_usage_gb") > 0, "Yes")
    .otherwise("No")
)
```

Part 7: Window Functions

```python
from pyspark.sql.window import Window

w1 = Window.orderBy(desc("data_used_gb"))
full_df = full_df.withColumn(
    "usage_rank",rank().over(w1)
)
full_df.show()
```

```python
w2 = Window.orderBy(desc("amount_paid"))
full_df = full_df.withColumn("payment_rank",rank().over(w2))
```

```python
full_df.orderBy(desc("data_used_gb")).show(3)
```

```python
full_df.orderBy(desc("amount_paid")).show(3)
```

```python
city_window = Window.partitionBy("city").orderBy(desc("data_used_gb"))

full_df.withColumn(
    "rank",
    row_number().over(city_window)
).filter(col("rank")==1).show()
```

```python
plan_window = Window.partitionBy("plan_name").orderBy(desc("data_used_gb"))

full_df.withColumn("rank",row_number().over(plan_window)).filter(col("rank")==1).show()
```

```python
month_window = Window.orderBy("bill_month")

monthly_rev = payments_clean.groupBy("bill_month").agg(sum("amount_paid").alias("monthly_revenue"))
monthly_rev = monthly_rev.withColumn("running_total",
    sum("monthly_revenue").over(month_window)
)
monthly_rev.show()
```

```python
usage_window = Window.partitionBy("customer_id") .orderBy("usage_month")
usage_lag = usage_clean.withColumn(
    "previous_usage",
    lag("data_used_gb").over(usage_window)
)
```

```python
usage_lead = usage_clean.withColumn("next_usage",
    lead("data_used_gb").over(usage_window)
)
```

```python
usage_lag.show()
usage_lead.show()
```

```python
usage_lag.filter(col("data_used_gb") > col("previous_usage")).show()
```

```python
full_df = full_df.withColumn(
    "usage_category",
    when(col("data_used_gb") >= 70,"Heavy User")
    .when(col("data_used_gb") >= 30,"Medium User")
    .otherwise("Low User")
)
full_df = full_df.withColumn(
    "payment_category",
    when(col("amount_paid") >= 1000,"High Payment")
    .when(col("amount_paid") >= 500,"Medium Payment")
    .otherwise("Low Payment")
)
full_df = full_df.withColumn(
    "churn_risk",
    when(
        (col("status") == "Inactive") |
        (col("payment_status") != "Success"),
        "High Risk"
    )
    .when(col("data_used_gb") < 15,"Medium Risk")
    .otherwise("Low Risk")
)
full_df = full_df.withColumn(
    "over_usage_gb",
    col("data_used_gb") - col("data_limit_gb")
)
full_df = full_df.withColumn(
    "over_usage_flag",
    when(col("over_usage_gb") > 0,"Yes")
    .otherwise("No")
)
```
