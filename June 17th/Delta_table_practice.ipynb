```python
# Create sample data
data = [
    (101, "Rahul", 75000),
    (102, "Priya", 85000),
    (103, "Amit", 65000)
]

# Create DataFrame
df = spark.createDataFrame(
    data,
    ["emp_id", "name", "salary"]
)

# Define volume path
volume_path = "/Volumes/hexa_databricks_7405610905139839/default/hexawarelake13/employees_delta"

# Write Delta files
df.write \
    .format("delta") \
    .mode("overwrite") \
    .save(volume_path)

print("Data written successfully")

# Read Delta files
delta_df = spark.read \
    .format("delta") \
    .load(volume_path)

# Display data
display(delta_df)
```
