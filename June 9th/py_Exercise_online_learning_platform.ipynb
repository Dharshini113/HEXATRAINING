```python
print("Total Activities:", rdd.count())
print("Unique Activities:", rdd.distinct().count())
print("Most Common Activity:",max(rdd.countByValue().items(),key=lambda x: x[1]))
```

```python
rdd.distinct().saveAsTextFile( "distinct_activities")
```

```python
rdd.map(lambda x: (x, 1)).reduceByKey(lambda a, b: a + b).saveAsTextFile("activity_frequencies")
```

```python
print("--LEARNING PLATFORM SUMMARY--")

most = max(act_cnts.items(), key=lambda x: x[1])
least = min(act_cnts.items(), key=lambda x: x[1])

print("=== Learning Platform Usage Summary ===")
print("Total Activities:", rdd.count())
print("Unique Activities:", rdd.distinct().count())
print("Most Common Activity:", most)
print("Least Common Activity:", least)
print("LOGIN Percentage:",(act_cnts['LOGIN'] / rdd.count()) *100)
print("VIDEO_PLAY Percentage:",(act_cnts['VIDEO_PLAY'] / rdd.count()) * 100)
print("Activity Frequencies:", act_cnts)
```
