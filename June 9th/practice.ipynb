```python
from pyspark.sql import SparkSession
spark = SparkSession.builder.appName('RDD').getOrCreate()
```

```python
numbers = [10,20,30,40,50]
rdd = spark.sparkContext.parallelize(numbers)
print(rdd.collect())
```

```python
rdd.count()
```

```python
rdd.take(3)
```

```python
result = rdd.map(lambda x: x * 10)
print(result.collect())
```

```python
result = rdd.map(lambda x: x * x)
print(result.collect())
```

```python
numbers = [1,2,2,3,4,5,6,7,7,8,9,10]
rdd = spark.sparkContext.parallelize(numbers)
result = rdd.filter(lambda x: x % 2 == 0)
print(result.collect())
```

```python
result = rdd.distinct()
print(result.collect())
```

```python
rdd1 = spark.sparkContext.parallelize([10,20,30])
rdd2 = spark.sparkContext.parallelize([40,50,60])
result = rdd1.union(rdd2)
print(result.collect())
```

```python
rdd = spark.sparkContext.parallelize([10,40,30,20])
result = rdd.sortBy(lambda x: x)
print(result.collect())
```

```python
result = rdd1.reduce(lambda x,y: x + y)
print(result)
```
