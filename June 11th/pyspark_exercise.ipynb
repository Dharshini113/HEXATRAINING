Part 7 – Window Functions

```python
from pyspark.sql.window import Window
```

```python
spending = patients_df.join(appointments_df, "patient_id") \
    .groupBy("patient_id","patient_name","city") \
    .agg(sum("consultation_fee").alias("total_fee"))
```

```python
w = Window.orderBy(desc("total_fee"))
```

```python
spending.withColumn("rank", rank().over(w)).show()
```

```python
spending.withColumn("dense_rank", dense_rank().over(w)).show()
```

```python
spending.withColumn("row_num", row_number().over(w)).show()
```

```python
spending.withColumn( "rank",rank().over(w)
).filter( "rank = 1").show()
```

```python
spending.orderBy(desc("total_fee")).show(1)
```

```python
spending.orderBy(desc("total_fee")).show(3)
```

```python
city_w = Window.partitionBy("city").orderBy(desc("total_fee"))
spending.withColumn("rank", rank().over(city_w)) \
     .filter(col("rank")==1).show()
```

```python
fee_w = Window.orderBy("total_fee")
```

```python
spending.withColumn(
    "running_total",
    sum("total_fee").over(fee_w)
).show()
```

```python
spending.withColumn(
    "lead_fee",
    lead("total_fee").over(fee_w)
).show()
```

```python
spending.withColumn(
    "lag_fee",
    lag("total_fee").over(fee_w)
).show()
```

Part 10 – ETL Project

```python
patients_df.show()
appointments_df.show()
preferences_df.show()
```

```python
patients_df = patients_df.fillna({
    "city":"Unknown",
    "blood_group":"Not Available"
})

appointments_df = appointments_df.fillna({
    "consultation_fee":0
})

flat_df = flat_df.na.fill({
    "phone":"Not Provided",
    "email":"Not Provided"
})
```

```python
final_df = patients_df.join(appointments_df,"patient_id","left") \
                   .join(preferences_df,"patient_id","left")
final_df.show()
```

```python
final_df = final_df.withColumn(
    "age_group",
    when(col("age") < 30,"Young")
    .when(col("age") < 50,"Adult")
    .otherwise("Senior")
)
final_df.show()
```

```python
patient_spending = final_df.groupBy(
    "patient_id","patient_name"
).agg(sum("consultation_fee").alias("total_spending"))
patient_spending.show()
```

```python
dept_revenue = final_df.groupBy(
    "department"
).agg(sum("consultation_fee").alias("department_revenue"))
dept_revenue.show()
```

```python
final_df.write.mode("overwrite").parquet("hospital_analytics")
```
