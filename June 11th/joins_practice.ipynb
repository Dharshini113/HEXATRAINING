```python
from pyspark.sql import SparkSession
spark = SparkSession.builder.appName("Working with Joins").getOrCreate()
```

```python
customers = [

(1, "Rahul", "Hyderabad"),
(2, "Priya", "Bangalore"),
(3, "Amit", "Mumbai"),
(4, "Sneha", "Chennai"),
(5, "Farhan", "Delhi")

]

customer_columns = [
    "customer_id",
    "customer_name",
    "city"
]

customers_df = spark.createDataFrame(
    customers,
    customer_columns
)

customers_df.show()
```

```python
orders = [

(101, 1, "Laptop", 65000),
(102, 2, "Mobile", 25000),
(103, 1, "TV", 45000),
(104, 3, "Chair", 5000),
(105, 7, "Watch", 8000)

]

order_columns = [
    "order_id",
    "customer_id",
    "product",
    "amount"
]

orders_df = spark.createDataFrame(
    orders,
    order_columns
)

orders_df.show()
```

```python
customers_df.join(orders_df,"customer_id","inner").show()
```

```python
customers_df.join(orders_df,"customer_id","left").show()
```

```python
customers_df.join(orders_df,"customer_id","right").show()
```

```python
customers_df.join(orders_df,"customer_id","full").show()
```

```python
from pyspark.sql.functions import *

customers_df.join(orders_df,"customer_id")\
.groupBy("customer_id","customer_name")\
.agg(sum("amount").alias("total_amount"))\
.show()
```

Window Functions


```python
employees = [

(101,"Rahul","IT",75000),
(102,"Priya","IT",85000),
(103,"Amit","IT",65000),

(104,"Sneha","HR",70000),
(105,"Farhan","HR",90000),

(106,"Neha","Finance",95000),
(107,"Arjun","Finance",80000),
(108,"Meera","Finance",75000)

]

columns = [
    "employee_id",
    "employee_name",
    "department",
    "salary"
]

df = spark.createDataFrame(
    employees,
    columns
)

df.show()
```

```python
from pyspark.sql.window import Window

window_spec = Window.orderBy(
    col("salary").desc()
)
df.withColumn(
    "row_number",row_number().over(window_spec)
    ).show()
```

```python
window_spec = Window.orderBy(col("salary").desc())
df.withColumn(
"rank",
rank().over(window_spec)
).show()
```

```python
window_spec = Window.orderBy(
col("salary").desc()
)
df.withColumn(
"dense_rank",
dense_rank().over(window_spec)
).show()
```

```python
window_spec = Window.partitionBy("department").\
orderBy(col("salary").desc())

df.withColumn("department_rank",rank().over(window_spec)).show()
```
